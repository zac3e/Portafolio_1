```python
def Graficar_delitos(Delito, Sub_deli, Entidad, Ano):
    df = data[(data["Entidad"] == str(Entidad)) & (data["Año"] == Ano) & (data["Tipo de delito"] == str(Delito)) & (data["Subtipo de delito"] == str(Sub_deli))]
    meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
    eje_X = df[meses].T
    plt.figure(figsize=(10, 6))
    plt.plot(eje_X)
    plt.xlabel('Mes')
    plt.ylabel('Cantidad de Delitos')
    plt.title(f'Cantidad de Delitos de {Delito} - {Sub_deli} en {Entidad} durante {Ano}')
    plt.xticks(range(len(meses)), meses, rotation=45)
    plt.grid(True)
    plt.show()
```

```python
Graficar_delitos("Delito sexual", "Abuso sexual", "Zacatecas", 2020)
```

```python
def Graficar_delitos_porano(Delito, Sub_deli, ano):
  df = data[(data["Año"] == ano) & (data["Tipo de delito"] == str(Delito)) & (data["Subtipo de delito"] == str(Sub_deli))]
  x=df["Entidad"].unique().tolist()
  meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
  sum=df[meses].sum(axis=1)
  plt.figure(figsize=(10, 6))
  plt.xticks(range(len(x)), x, rotation=90)
  plt.ylabel("Cantidad de delitos")
  plt.title(f"Cantidad de delitos de {Delito} - {Sub_deli} por estado en el año {ano}")
  plt.bar(x,sum )
```

```python
Delito="Robo con violencia"
Sub_deli="Robo a transportista"
ano=2020
Graficar_delitos_porano(Delito, Sub_deli, ano)
```

```python
Delito="Fraude"
Sub_deli="Fraude"
ano=2020
Graficar_delitos_porano(Delito, Sub_deli, ano)
```
