##Detección de Spam en Correo Electrónico
El proposito de este programa es entrenar un modelo de naiver bayes para la detección de spam con el dataset descrito a continucaicon

El archivo csv contiene 5172 filas, cada fila para cada correo electrónico. Hay 3002 columnas. La primera columna indica el nombre del correo electrónico. El nombre se ha configurado con números y no con el nombre de los destinatarios para proteger la privacidad. La última columna tiene las etiquetas de predicción: 1 para spam, 0 para no spam. Las 3000 columnas restantes son las 3000 palabras más comunes en todos los correos electrónicos, después de excluir los caracteres/palabras no alfabéticos. Para cada fila, el recuento de cada palabra (columna) en ese correo electrónico (fila) se almacena en las celdas respectivas. Por lo tanto, la información sobre todos los correos electrónicos 5172 se almacena en un marco de datos compacto en lugar de como archivos de texto separados.

## Notemos que el codigo será muy compacto pues solo se trata en este caso de aplicar el modelo, pues el conjunto de datos esta muy bien preparado

```python
for i in df.columns.values.tolist():
  if df[i].isnull().sum() != 0: #No devuelve nada y eso es bueno, pues cumple la condicon que no hay datos null
    print(f"en la columna {i}  hay un dato null")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
columnas  = df.columns.values.tolist()
```

```python
columnas_2  =columnas.copy()
Y=df["Prediction"]
columnas_2.remove("Prediction")
columnas_2.remove("Email No.")
X=df[columnas_2 ]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.3, random_state=42
)
```

```python
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train, y_train)
```

```python
y_pred = nb_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
print(f"Exactitud: {accuracy:.2f}")
print("Reporte de clasificación:\n", report)
```

```python
Predicion=pd.DataFrame( nb_classifier.predict(X))
```

```python
Predicion["Y_original"]=Y
```
